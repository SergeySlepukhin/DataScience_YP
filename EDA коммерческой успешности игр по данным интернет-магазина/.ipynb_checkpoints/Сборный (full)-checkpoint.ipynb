Выше мы отметили, что для столбцов:

   - `user_score`: object целесообразно изменить на float64 (природа данного параметра количественная);

Для проверки наличия дробной части отличной от .0 в столбце `critic_score` напишем собственную функцию:

````python
df_score = data['critic_score']
df_score = df_score.fillna(0)
df_score = df_score.astype('float').to_frame()
````

````python
def decimal_part(n):
    try:
        b = n * 10
        c = b % 10
        d = c / 10
        e = round(d, 10)
        return e
    except:
        return n
````

````python
df_score = data['critic_score'] # копия столбца
df_score = df_score.fillna(0) # заменим пропуски на 0 в целях проверки
df_score = df_score.astype('float').to_frame()
df_score['decimal'] = df_score['critic_score'].apply(decimal_part) # применим функцию
df_score['decimal'].mean() # выведем ср. значение дробных частей в столбце 
````

Таким образом, имеет смысл заменить тип данных в `critic_score` c float64 на int

````python
data[['year_of_release', 'critic_score']] = data[['year_of_release', 'critic_score']].astype('Int64')
````

Перед изменением типа данных в столбце `user_score`, исследуем количество значений 'tbd', так как без их предобработки не получится произвести конвертацию в нужный тип данных

### Обработка пропусков и явных дубликатов

Визуализируем количество пропусков в столбцах с помощью собственной функции:

````python
def pass_value_barh(df):
    try:
        ax = \
        (
            (df.isna().mean()*100)
            .to_frame().rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
        )
        ax = ax.plot(kind = 'barh', figsize = (19,10), rot = 0, legend = True, fontsize = 16) 
        plt.title('Количество пропусков в столбцах (%)' + "\n", fontsize = 22, color = 'Black')
        plt.bar_label(ax.containers[0], fmt='%.2f%%', padding=0, color = 'red')
        plt.xlabel('проценты', fontsize=16)
        plt.xticks(np.arange(0, 100, step=5))

    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
````

1) пропуски в исследуемых столбцах нецелесообразно заполнять никакими синтетическими данными (средним, медианой и т.д.): учитывая объем пропусков (от 40 % до 55 %) и отсутствие явной логической связи со значениями из других столбцов, заполнение синтетическими данными приведет к существенным искажениям в первоначальной выборке;

2) учитывая корр. связь пропусков в исследуемых столбцах, количество пропусков, а также смысл рассматриваемых переменных (оценки пользователей, критиков, возрастные рейтинги), скорее всего, в большинстве случаем имело место воздействие внешних факторов: 
   - например, для  `critic_score`, `user_score` - реальное отсуствие таких рейтингов для не самых популярных (многие игры могли просто "не дойти" до критиков в силу их неизвестности, коммерческой неуспешности или небольшого потенциала) или условно "старых" игр, выпущенных в те времена, когда не велась статистика по scores;
   - что касается `rating` by ESRB:
     -возрастные рейтинги данной организации активно используются именно в Северной Америке;
     - в Европе более популярна Pan European Game Information (PEGI);
     - в Японии - Computer Entertainment Rating Organization (CERO).

````python
data['user_score'] = data['user_score'].fillna(-1)
data['critic_score'] = data['critic_score'].fillna(-1)
data['rating'] = data['rating'].fillna('unknown')
````

Действительно, коэффициент свидетельствует **о наличии или отсутствии линейной связи** между переменными, и **измеряет выраженность** этой связи. То есть, влияния оценок на продажи нет. Коэффициент корреляции Пирсона не про это. Он про линейную зависимость 2-х переменных. И ещё, это коэффициент **НЕ ОБЪЯСНЯЕТ** зависимость, он **НЕ ГОВОРИТ** почему наблюдается такой эффект, но просто констатирует факт, что корреляция наблюдается (например, "мы видим, что те игры, которым наши эксперты дают высокие оценки - продаются лучше". А вот почему они продаются лучше - можно только предполагать ... разные причины могут быть).</div>

Исходя из формулировки, задача состоит в сравнении средних двух генеральных совокупностей на основе имеющихся выборок. Соответственно нулевая гипотеза будет исходить из равенства 2-х генеральных совокупностей, альтернативная - из их неравенства.

```
H_0: Пользовательский рейтинг (user_score) платформы "Xbox One" = пользовательский рейтинг (user_score) платформы "PC"
H_a: Пользовательский рейтинг (user_score) платформы "Xbox One" ≠ пользовательский рейтинг (user_score) платформы "PC"`
alpha = 0.05
```

Срез пользовательских оценок для Xbox One и PC

````python
xbox_one = stream.query('(platform == "XOne") and (user_score != -1)')['user_score'].reset_index(drop=True)
pc = stream.query('(platform == "PC") and (user_score != -1)')['user_score'].reset_index(drop=True)
````

<div class="alert alert-success">
<b>ОТЛИЧНО! 👍</b>

Очень хорошо, что ты не забываешь отфильтровать «заглушки»
</div>


````python
xbox_one_var = xbox_one.var()
pc_var = pc.var()
dif = len(pc) / len(xbox_one)

print(f'Дисперсия XOne: {xbox_one_var:0.2}, Дисперсия PC: {pc_var:0.2}, разница в длине выборок - в {dif:0.2} раз')
````

Видим, что дисперсия отличается (хотя и несильно), выборки разного объема и разница существенна

---
посмотрим на распределение значений в выборках

````python
# check
xbox_one.hist(bins = 25, figsize = (15,3), range = (0,10), alpha=0.5, color='green')

pc.hist(bins = 25, figsize = (15,7), range = (0,10), alpha=0.5)

plt.legend([' Xbox One', 'PC'])
plt.xlabel('Оценка пользователей')
plt.ylabel('Количество игр')
plt.show()
````

Протяженный левый "хвост" навряд ли позовляет назвать распределения нормальными. **В качестве упражнения** выведем коэфициенты асимметрии и эксцесса - показатели отклонения от нормальности 

````python
from scipy.stats import skew, kurtosis

display(skew(xbox_one, bias=False)) # коэфициенты асимметрии 
skew(pc, bias= False )
````

<div class="alert alert-warning">
<b>Комментарий 👉</b>

Все импорты лучше собрать в самой первой ячейке "тетрадки" с кодом - так твои коллеги с первого взгляда смогут понять и оценить какие библиотеки ты используешь в своём проекте.</div>


````python
display(kurtosis(xbox_one, bias=False)) # коэфициенты эксцесса 
kurtosis(pc, bias= False )
````

Видим, что оба распределения обладают существенной левосторонней асимметрией + 'хвосты' "легче», а пик острее, чем у нормального распределения.

````python
from scipy.stats import normaltest
stat, p = normaltest(pc)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
````

````python
stat, p = normaltest(xbox_one)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
````

Статистическая функция normaltest (), основанная на оценке асимметрии и эксцесса, также отказывает в нормальности. 

**Ненормальность выборочного распределения не должна нас смущать!**
Так как наши выборки > 30 значений, ЦПТ, позволяет утверждать, что выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности.

---
**Соответственно есть все основания для использования t-test**

```
1) рейтинги пользователй для 2-х рассматриваемых платформ не зависят друг от друга;

2) каждая выборка состоит из более чем 30 значений;

3) выборки сильно отличаются по количеству значений, следовательно при выполнении t-test укажeм False для параметра `equal_var`; также (см. ниже) отличаются и дисперсии, однако для проверки значимости этого отличия нужно делать F-test, а отличия в размере выборок уже позволяют принять решение о значении параметра.
```

````python
xbox_one_score = xbox_one.tolist()
pc_score = pc.tolist()

results = results = st.ttest_ind(xbox_one_score, pc_score, equal_var=False)

alpha = 0.05 

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

<div class="alert alert-success">
<b>ОТЛИЧНО! 👍</b>

Верно то, что мы используем здесь двусторонний тест.

Ну и отлично, что ты понимаешь, что нельзя сделать вывод, что тест "подтвердил" одну из гипотез (в статистике гипотезы подтвердить невозможно) или "подтвердил" равенство/неравенство (можно говорить только о статистически (не)значимом различии). Например, так: «У нас нет оснований утверждать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются».

И ещё на эту же тему: никакие экспериментально полученные данные никогда не подтвердят какую-либо гипотезу. Это наше фундаментальное ограничение. Данные могут лишь не противоречить ей или, наоборот, показывать крайне маловероятные результаты (при условии, что гипотеза верна). Но и в том, и в другом случае нет оснований утверждать, что выдвинутая гипотеза доказана. Допустим, данные гипотезе не противоречат, тогда мы её не отвергаем. Если же мы приходим к выводу, что получить такие данные в рамках этой гипотезы вряд ли возможно, у нас появляется основание отбросить эту гипотезу. 

Статья на эту тему: https://allatambov.github.io/psms/pdf/hypo-test.pdf
</div>


**Отвергаем нулевую гипотезу:** средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

Видим, что существует вероятность 95%, что указанные доверительные интервалы содержат средние истинные оценки пользователей. При этом нижня граница дов. ин-ла для PC в любом случае больше верх. границы дов. интервала для XOne

### Средние пользовательские рейтинги жанров Action и Sports разные

Исходя из формулировки, задача состоит в сравнении средних двух генеральных совокупностей на основе имеющихся выборок. Соответственно нулевая гипотеза будет исходить из равенства 2-х генеральных совокупностей, альтернативная - из их неравенства.

```
H_0: Пользовательский рейтинг (user_score) игр жанра Action = Пользовательский рейтинг (user_score) игр жанра Sports
H_a: Пользовательский рейтинг (user_score) игр жанра Action ≠ Пользовательский рейтинг (user_score) игр жанра Sports
alpha = 0.05
```

````python
action = stream.query('(genre == "Action") and (user_score != -1)')['user_score'].reset_index(drop=True)
sports = stream.query('(genre == "Sports") and (user_score != -1)')['user_score'].reset_index(drop=True)
````

````python
action_var = action.var()
sports_var = sports.var()
dif = len(action) / len(sports)

print(f'Дисперсия Action: {action_var:0.2}, Дисперсия Sports: {sports_var:0.2}, разница в длине выборок - в {dif:0.2} раз')
````

Размеры выборок также существенно отличаются + разница дисперсий в распределении значений является еще более существенной, чем для рассмотренных платформ.

````python
# check
action.hist(bins = 25, figsize = (15,3), range = (0,10), alpha=0.5, color='green')

sports.hist(bins = 25, figsize = (15,7), range = (0,10), alpha=0.5)

plt.legend(['Action', 'Sports'])
plt.xlabel('Оценка пользователей')
plt.ylabel('Количество игр')
plt.show()
````

Распределение пользовательских оценок по жанрам аналогично распределению по платформам имеет длинный левый "хвост". По причинам изложенным выше нас это смущать не должно, воспользуемся **t-test**: в данном случае всего 3 условия также выполняются.

````python
action_score = action.tolist()
sports_score = sports.tolist()

results = results = st.ttest_ind(action_score, sports_score, equal_var=False)

alpha = 0.05 

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Отвергаем нулевую гипотезу:** средние пользовательские рейтинги для игр рассмотренных разных жанров действительно отличаются между собой. 

**Построим 95% доверительные интервалы**

````python
display(action.mean())

st.norm.interval(alpha=0.95, loc=np.mean(action), scale=st.sem(action)) 
````

````python
display(sports.mean())

st.norm.interval(alpha=0.95, loc=np.mean(sports), scale=st.sem(sports)) 
````

Видим, что существует вероятность 95%, что указанные доверительные интервалы содержат средние истинные оценки пользователей. При этом нижняя граница дов. ин-ла для Action в любом случае больше верх. границы дов. интервала для Sports.

Возможно, что рамки каждого из этих жанров ограничивают разнообразие выпускаемых игр, при этом **коммерческая эффективность такого жанра как Platform говорит о том, что величина предложения в рамках одного жанра (количество выпускаемых игр) != коммерческая успешность отдельно взятой игры**: так, наиболее многочисленный жанр Action имеет мединаную стоимсть всего в 110 тыс. рублей.
