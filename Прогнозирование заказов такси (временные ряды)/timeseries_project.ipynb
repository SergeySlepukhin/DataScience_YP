#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

# Цель исследования
построить оптимальную модель для задачи регрессии временного ряда, которая будет способна спрогнозировать на тестовой выборке примерную количество заказов такси на следующий час по данным о пред. по времени заказах; оптимальность модели определяется в соответствии с критериями заказчика.

## Анализ

### Тренды и сезонность

С помощью функции `seasonal_decompose()` разложим наш временной ряд на:
- trend — тренд;
- seasonal — сезонную составляющую;
- residuals — остатки декомпозиции.

```python
decomposed = seasonal_decompose(df)
decomposed_days = seasonal_decompose(df_days)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend_hourly')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality_hourly')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals_hourly')
plt.tight_layout()

plt.figure(figsize=(12, 8))
plt.subplot(311)
decomposed_days.trend.plot(ax=plt.gca())
plt.title('Trend_daily')
plt.subplot(312)
decomposed_days.seasonal.plot(ax=plt.gca())
plt.title('Seasonality_daily')
plt.subplot(313)
decomposed_days.resid.plot(ax=plt.gca())
plt.title('Residuals_daily')
plt.tight_layout()
```

**Декомпозиция ряда, ресемплированного по часам**, дала мало информации: разве что можно выделить восходящий, видимо, годовой тренд, характеризующий, как минимум, общий рост заказов такси в весной-летом 2018.
 
**Декомпозиция ряда, ресемплированного по дням более информативна:**
- линейный восходящий тренд, с небольшими "провалами" в начале каждого месяца (кроме июля) и ускоренным ростом в августе, стал более очевидным;
- можно выделить недельную сезонную составляющую: в каждом месяце прослеживается регулярный период изменений (устойчивые колебания 2-х минимумов и 2-х максимумов), который повторяется 4 раза, что соответствует, видимо, 4-м неделям;
- остатки с большой дисперсией относительно 0, что говорит о неустойчивости периодических изменений ряда
---

**Теперь построим графики временного ряда, ресемплированные по часам и дням, более "близко":**
- отобразим на **графике ряда, ресемплированного по часам**, скользящее среднее и стандартное отклонение за 24 часа и неделю (168 часов), исходя из гипотезы о наличии суточной сезонности и возрастающего линейного тренда

- отобразим на **графике ряда, ресемплированного по дням**, скользящее среднее и стандартное отклонение за 7 дней и месяц, исходя из гипотезы о наличии недельной сезонности и возрастающего линейного тренда

```python
plt.figure(figsize=(15,8))
 
plt.plot(df, label = 'заказы такси по часам', color = 'steelblue')
# сглаживаем по дням
plt.plot(df.rolling(window=24).mean(), label = 'cкользящее среднее за 24 часа', color = 'orange')
plt.plot(df.rolling(window=24).std(), label = 'cкользящее стандартное отклонение за 24 часа', color = 'green')
# сглаживаем по неделям
plt.plot(df.rolling(window=168).mean(), label = 'cкользящее среднее за неделю', color = 'red')
plt.plot(df.rolling(window=168).std(), label = 'cкользящее стандартное отклонение за неделю', color = 'black')

plt.legend(title='', loc='upper left', fontsize=14)
plt.xlabel('Месяц', fontsize = 14)
plt.ylabel('Количество заказов', fontsize = 14)
plt.title('Заказы такси с марта по август 2018', fontsize = 25)
plt.show()

plt.figure(figsize=(15,8))
 
# поочередно зададим кривые (перевозки и скользящее среднее) с подписями и цветом
plt.plot(df_days, label = 'заказы такси по дням', color = 'steelblue')
plt.plot(df_days.rolling(window=7).mean(), label = 'cкользящее среднее за 7 дней', color = 'orange')
plt.plot(df_days.rolling(window=7).std(), label = 'скользящее стандартное отклонение за 7 дней', color = 'green')
plt.plot(df_days.rolling(window=30).mean(), label = 'скользящее среднее за 14 дней', color = 'red')
plt.plot(df_days.rolling(window=30).std(), label = 'скользящее стандартное отклонение за 14 дней', color = 'black')

 
# добавим легенду, ее положение на графике и размер шрифта
plt.legend(title='', loc='upper left', fontsize=14)
 
# добавим подписи к осям и заголовки
plt.xlabel('Месяц', fontsize = 14)
plt.ylabel('Количество заказов', fontsize = 14)
plt.title('Заказы такси с марта по август 2018', fontsize = 25)
 
# выведем обе кривые на одном графике
plt.show()
```

**Видим:**
- скользящее среднее на обих графиках хорошо визуализирует **слабый восходящий тренд, который в августе 2018 усиливается и становится фактически нелинейным**;
- при этом скользящее стандартное отклонение на протяжении всего периода, кроме последнего месяца (в августе возрастает), является стабильным
---
таким образом, **визуально ряд оценивается как "околостационарный"**
- график временного ряда, ресемплированный по часам, по-прежнему является перегруженным: пока наличие/ отсутствие суточной сезонности оценить сложно;
- график временного ряда, ресемплированный по дням, допольнительно (к результатам автоматической декомпозиции) дает основания предполагать недельную сезонность:
    - визуально можно оценить регулярный период изменений (регулярно повторяющаяся последовательность 2 максимумов и 2 минимумов, не всегда прослеживается четко: наиболее четко в первые месяцы, в летние, особенно в августе - наименее четко, видимо, влияют также другие факторы), который повторяется 4 раза в мес, что, в целом, соответствует неделе.

Для выявления **трендов наиболее интересного для нас месяца - августа и суточной (ежедневной) сезонности в этом месяце декомпозируем его**

```python
decomposed_august = seasonal_decompose(df['2018-08-01':'2018-08-31'])
plt.figure(figsize=(12, 8))
plt.subplot(311)

decomposed_august.trend.plot(ax=plt.gca())
plt.title('Trend_august')
plt.subplot(312)
decomposed_august.seasonal.plot(ax=plt.gca())
plt.title('Seasonality_august')
plt.subplot(313)
decomposed_august.resid.plot(ax=plt.gca())
plt.title('Residuals_august')
plt.tight_layout()
```

Видим:
- восходящие и нисходящие тренды августа подчеркивают еженедельную сезонность:
    - пики вс-пн + между ними локальные пики, соответствующие, скорее всего, пятницам
- ежесуточная сезонность показывает, что суточный тренд характеризуется глобальным (в пределах суток) повышением количества заказов такси, достигая пика, видимо, ближе к ночи, день характеризуется уменьшением и увеличением заказов относительно 2-х локальных пиков;
- остатки показывают, что заказы с вечера вс по ночь пн очень выбиваются из общего ряда, прослеживатеся автокорреляция остатков...

**Посмотрим поближе: возьмем срезы - одну из первых недель (марта-апрель), одну из последних недель в августе, таким образом, чтобы захватить все дни недели.**

```python
start_0, end_0 = '2018-03-25', '2018-04-03'
start_1, end_1 = '2018-08-12', '2018-08-21'
```

### Исследование стационарности ряда

Выше мы заметили, что среднее значение до июля и дисперсия на всем периоде наблюдений не особо меняются со смещением во времени.
Выдвинув гипотезу о том, что ряд является околостационарным. Проверим ее статистическую значимость с помощью теста **Дики-Фуллера (Dickey-Fuller test)**.

`H_0` : временной ряд является нестационарным. Другими словами, он имеет некоторую структуру, зависящую от времени, и не имеет постоянной дисперсии во времени.

`H_A` : временной ряд является стационарным.

```python
# импортируем необходимую функцию 
from statsmodels.tsa.stattools import adfuller
 
# передадим ей df с одним столбцом
adf_test = adfuller(df)
 
# выведем p-value
print('p-value = ' + str(adf_test[1]))
```

p-значение < 0,05, мы можем отвергнуть нулевую гипотезу; с другой стороны при уровне значимости < 0.028, мы бы ее не отвергли.
Будем считать наш ряд почти стационарным; видимо, августовские значения тест посчитал, скорее, случайными и маловероятными 

**Избавиться от "остаточной" стационарности, как известно, можно с помощью разницы временного ряда и его "сдвига"**

```python
df_stationary = df - df.shift()
df_stationary['mean'] = df_stationary['num_orders'].rolling(15).mean()
df_stationary['std'] = df_stationary['num_orders'].rolling(15).std()
df_stationary.plot()
```

```python
df_stationary = df_stationary.drop(['mean','std'], axis=1)
df_stationary = df_stationary.dropna()
df_stationary.head()
```

```python
adf_test = adfuller(df_stationary)
 
# выведем p-value
print('p-value = ' + str(adf_test[1]))
```

Действительно, ряд стал однозначно стационарным!

### Создание признаков

Напишем собственную функцию для автоматизации создани признаков для обучения прогнозной модели по заданным параметрам

```python
def make_features(data, max_lag, rolling_mean_size):
    # calendar features
#     data['month'] = data.index.month
#     data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    data['lag_168'] =data['num_orders'].shift(168)
    #lag features fitting
    for i in range(1, max_lag + 1):
        data['lag_' + str(i)] = data['num_orders'].shift(i)
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data['std'] = data['num_orders'].shift().rolling(rolling_mean_size).std()
    return data
```

```python
df_learning = df.copy() 
```

создадим таблицу, в которой будем сохранять важные для заказчика критерии 

```python
model_results = pd.DataFrame(
    columns = [
        'Модель',
        'Время обучения (s.)',
        'Время предсказания (s.)',
        'RMSE валидация',
        'RMSE тест'
    ])
model_results
```

Подберем наиболее оптимальные аргументы функции (максимальный лаг и размер скользящего среднего) обучив базовую модель линейной регресии; задача оптимизации в данном случае - минимизация метрики RMSE. 

```python
# def RMSE(target, predict):
#     RMSE = mean_squared_error(target, predict) ** 0.5
#     return RMSE

# score = make_scorer(RMSE, greater_is_better=False)
```

**с помощью циклов автоматизируем подбор аргументов функции `make_features`**

---
оценивать качество прогноза в зависимости от аргументов функции будем на последовательно выделенной валидационной выборке (10% от исходной выборки)

```python
%%time
best_rmse = 50
for lag in tqdm(range(24, 50)):
    for rolling_mean in tqdm(range(2, 10)):
        df_lr_find = make_features(df_learning, lag, rolling_mean)
        train_lr, test_lr = train_test_split(df_lr_find, shuffle=False, test_size=0.2)
        valid_lr, test_lr = train_test_split(test_lr, shuffle=False, test_size=0.5)
        train_lr = train_lr.dropna()
        features_train = train_lr.drop("num_orders", axis=1)
        target_train = train_lr["num_orders"]

        features_valid = valid_lr.drop("num_orders", axis=1)
        target_valid = valid_lr["num_orders"]
        

        LR = LinearRegression()
        LR.fit(features_train, target_train)

        preds = LR.predict(features_valid)

        rmse = RMSE(preds, target_valid)
        if rmse < best_rmse:
            best_rmse = rmse
            best_lag = lag
            best_rolling_mean = rolling_mean
            print('RMSE:', best_rmse,
                  'лаг:', best_lag,
                  'скользящее среднее:', best_rolling_mean)

print("The best RMSE:", best_rmse, "\nThe best rolling_mean:", best_rolling_mean, "\nThe best lag:", best_lag)
```
